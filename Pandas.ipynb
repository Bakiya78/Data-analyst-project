```python
import pandas as pd
```

```python
data = pd.read_csv("Dataset.csv")
```

```python

data['total_spent'] = data.groupby('id')['income'].transform('sum')
data['purchase_frequency'] = data.groupby('id')['membership_years'].transform('count')
print(data['total_spent'] )
print(data['purchase_frequency'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[['total_spent', 'purchase_frequency',]] = scaler.fit_transform(data[['total_spent', 
                                                                                    'purchase_frequency']])
```

```python
print(data.head())
print(data.tail())
print(data[['total_spent', 'purchase_frequency']] )
```

```python
print(data.shape)
print(data.isnull() .sum())
```

```python
data.dropna(inplace=True)
print(data.info())
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=0)
data['cluster'] = kmeans.fit_predict(data[['total_spent', 'purchase_frequency']])
print(data['cluster'] )
```

```python
data.to_csv("Dataset.csv", index=False)
data.drop(columns = "Unnamed: 9")
```
